```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as date
```

```python
df_tweetSentiment = pd.read_csv('Data/SampleTweets.csv', names=['Sentiment', 'Date', 'Tweet', 'Value'])
df_stockPrices = pd.read_csv('Data/StockPrices.csv', names = ['Date', 'Open', 'High', 'Low'])
df_tweetSentiment.Date = pd.to_datetime(df_tweetSentiment.Date)
df_stockPrices.Date = pd.to_datetime(df_stockPrices.Date);
```

```python
df = pd.merge(df_stockPrices, df_tweetSentiment, on = 'Date')
```

```python
df.head(500)
```

```python
df = df.groupby('Date').mean().reset_index()
```

```python
x = df.Date
y1 = df.High
y2 = df.Value

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y1, 'r-')
ax2.plot(x, y2, 'g-')

ax1.set_xlabel('Date')
ax1.set_ylabel('High', color='r')
ax2.set_ylabel('Value', color='g')

plt.title('High Price vs. Value')

plt.show()
```

```python
x = df.Date
y1 = df.Low
y2 = df.Value

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y1, 'r-')
ax2.plot(x, y2, 'g-')

ax1.set_xlabel('Date')
ax1.set_ylabel('Low', color='r')
ax2.set_ylabel('Value', color='g')

plt.title('Low Price vs. Value')

plt.show()
```

```python
x = df.Date
y1 = df.Open
y2 = df.Value

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y1, 'r-')
ax2.plot(x, y2, 'g-')

ax1.set_xlabel('Date')
ax1.set_ylabel('Open', color='r')
ax2.set_ylabel('Value', color='g')

plt.title('Open Price vs. Value')

plt.show()
```
